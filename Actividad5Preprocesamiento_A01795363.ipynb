5. ¿Tendrán los valores faltantes de `Weight` relación con alguna otra variable? Analízalo considerando las siguientes columnas:

Para ello deberás dividir cada columna en dos grupos: con/sin faltantes en `Weight` y utilizar métodos gráficos y la prueba t para ver si hay diferencias significativas.

**NOTA**. Puedes reutilizar las funciones `Diagnose_MV_Numerical` y `Diagnose_MV_Categorical` definidas en Hands-On-Prep_Missing_Values.zip. Revisa nuevamente cómo utilizarlas.

```python
from scipy.stats import ttest_ind
def Diagnose_MV_Numerical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    labels=[]
    box_sr = pd.Series('',index = BM_MV.unique())
    for poss in BM_MV.unique():
        BM = BM_MV == poss
        box_sr[poss] = df[BM][str_att_name].dropna()
        labels.append(MV_labels[poss])

    plt.boxplot(box_sr,vert=False)
    plt.yticks([1,2],labels)
    plt.xlabel(str_att_name)
    plt.show()

    plt.figure(figsize=(10,4))

    att_range = (df[str_att_name].min(),df[str_att_name].max())

    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].hist()
        plt.xlim = att_range
        plt.xlabel(str_att_name)
        plt.title(MV_labels[poss])

    plt.show()

    group_1_data = df[BM_MV][str_att_name].dropna()
    group_2_data = df[~BM_MV][str_att_name].dropna()

    p_value = ttest_ind(group_1_data,group_2_data).pvalue

    print('p-value of t-test: {}'.format(p_value))
```

```python
BM_MV = df['Weight'].isnull()  # Boolean mask: True for missing values, False otherwise
Diagnose_MV_Numerical(df, 'Age', BM_MV)
```

```python
Diagnose_MV_Numerical(df, 'Number of siblings', BM_MV)
```

```python
from scipy.stats import chi2_contingency
def Diagnose_MV_Categorical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    plt.figure(figsize=(10,4))
    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].value_counts().plot.bar()
        plt.title(MV_labels[poss])
    plt.show()

    contigency_table = pd.crosstab(BM_MV,df[str_att_name])
    p_value = chi2_contingency(contigency_table)[1]

    print('p-value of Chi_squared test: {}'.format(p_value))
```

```python
BM_MV = df['Weight'].isnull()  # Boolean mask: True for missing values, False otherwise
Diagnose_MV_Categorical(df, 'Gender', BM_MV)
```

7. ¿Hay valores atípicos de la variable `Weight`? Confírmalo con un gráfico de boxplot. ¿Cuántos son? Almacénalos en un dataframe independiente y elimínalos del conjunto original.

```python
import matplotlib.pyplot as plt

# 1. Graficar un boxplot de 'Weight'
plt.boxplot(df['Weight'].dropna())  # Eliminar valores nulos para el gráfico
plt.title("Boxplot de Weight")
plt.ylabel("Weight")
plt.show()

# 2. Identificar y contar los valores atípicos
Q1 = df['Weight'].quantile(0.25)
Q3 = df['Weight'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites para considerar un valor como atípico
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los valores atípicos
outliers = df[(df['Weight'] < lower_bound) | (df['Weight'] > upper_bound)]

# 3. Almacenar los valores atípicos en un DataFrame independiente
df_outliers = outliers.copy()

# Mostrar la cantidad de valores atípicos
print(f"Total de valores atípicos: {len(df_outliers)}")

# 4. Eliminar los valores atípicos del DataFrame original
df = df.drop(outliers.index)

# Verificar
print(df['Weight'].describe())
```

9. ¿Hay encuestados que sean atípicos según su índice de masa corporal (`BMI`)? El `BMI` puede ser calculado usando la siguiente fórmula:

`BMI = Weight / Height ^ 2`

**NOTA**. El peso tiene que estar en kilogramos y la altura en metros, pero en el conjunto la altura se ha registrado en centímetros, por lo que deberás convertirla a metros.

```python
# Convertir la altura de centímetros a metros
df['Height_m'] = df['Height'] / 100

# Calcular el BMI
df['BMI'] = df['Weight'] / df['Height_m']**2

# Graficar un boxplot de BMI
plt.boxplot(df['BMI'].dropna())  # Eliminar valores nulos para el gráfico
plt.title("Boxplot de BMI")
plt.ylabel("BMI")
plt.show()

# Identificar y contar los valores atípicos de BMI usando IQR
Q1_BMI = df['BMI'].quantile(0.25)
Q3_BMI = df['BMI'].quantile(0.75)
IQR_BMI = Q3_BMI - Q1_BMI

# Definir los límites para considerar un valor como atípico
lower_bound_BMI = Q1_BMI - 1.5 * IQR_BMI
upper_bound_BMI = Q3_BMI + 1.5 * IQR_BMI

# Filtrar los valores atípicos de BMI
outliers_BMI = df[(df['BMI'] < lower_bound_BMI) | (df['BMI'] > upper_bound_BMI)]

# Mostrar la cantidad de valores atípicos
print(f"Total de encuestados atípicos según BMI: {len(outliers_BMI)}")
```
